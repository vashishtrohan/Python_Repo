```python
##### import os
username_password= 'username_password.txt'

def register():
    username= input("enter the username")
    password= input("enter the password")
    if username_exists(username):
        return
    with open(username_password, 'a') as f:
        f.write(f"{username},{password}\n")
    open (f"{username}_txt", 'a').close()
    print("successğŸ˜Š user has been registered succesfully")


def login():
    username=input("enter username")
    password= input("enter password")
    with open(username_password, 'r')as f:
        uname_pwd=f.readlines()
        for l in uname_pwd:
            uname, pwd=l.strip().split(',')
            if username==uname and password==pwd:
                print("login successfull")
                return username
        print("invalid username or password")
        return None
                

def username_exists(username):
    if not os.path.exists(f"{username}_txt"):
        return False
    else:
        print("user already exists")
        return True
    

def add_task(username):
    task=input("enter task to be completed")
    with open(f"{username}_txt", 'a') as f:
        f.write(f"{task}|0\n")
    print("task added successfully")


def view_task(username):
    with open (f"{username}_txt", 'r') as f:
        line_task=f.readlines()
        for l in line_task:
            desc, status=l.strip().split('|')
            status_desc= "completed âœŒğŸ¼" if status=='1' else "pending â°"
            print(f"{desc}-[{status_desc}]")
        
            
def marked_complete(username):
    with open(f"{username}_txt", 'r') as f:
        l_tasks=f.readlines()
        for i, desc in enumerate(l_tasks,1):
            desctask, status= desc.strip().split('|')
            desc_status="marked complete" if status==1 else "pending"
            print(f"{i}-{desctask}-[{desc_status}]")
            
    task_comp_index=int(input("what is the number of task you want to mark compelte"))-1
    with open(f"{username}_txt", 'r') as f:
        list_task=f.readlines()
       
        if 0<=task_comp_index<len(list_task):
            for index, desctask in enumerate(list_task):
                desc, status= desctask.strip().split('|')
                if index==task_comp_index:
                    list_task[task_comp_index]=f"{desc}|1\n"
                
        else:
            print("enter a valid task number to be mrked deleted")
    with open(f"{username}_txt", 'w') as f:
        f.writelines(list_task)
        print("task has been marked for deletion")
              
    

def delete_task(username):
    
    with open(f"{username}_txt", 'r') as f:
        l_tasks=f.readlines()
        for i, desc in enumerate(l_tasks,1):
            print(f"{i}-{desc}")
    delete_index=int(input("what task you want to delete give a number"))-1
    with open(f"{username}_txt", 'r') as f:
        l_tasks=f.readlines()
        l_tasks.pop(delete_index)
    with open(f"{username}_txt", 'w') as f:
        f.writelines(l_tasks)
    print("tasks deleted successfully")
        

def menu(username):
    while True:
        print("""Task Menu-
              1. add a task
              2. view tasks
              3. mark a task complete
              4. delete a task
              5. Logout
            """)
        choice= input("enter a choice")
        if choice=="1":
            add_task(username)
        elif choice=="2":
            view_task(username)
        elif choice=="3":
            marked_complete(username)
        elif choice=="4":
            delete_task(username)
        elif choice=="5":
            break
        else:
            print("enter correct number of choice")
        

    

def main():
    while True:
        print(""" Task Manager
                1. register a user
                2. login 
                3. Exit
                """)
        choice=input(" enter a choice")
        if choice=="1":
           register()
        elif choice=="2":
           user=login()
           if user:
               menu(user)   
        elif choice=="3":
            print("goodbye")
            break
        else:
            print("enter a correct choice")
            
    
    
    
if __name__=="__main__":
    main()
```
